### Load Packages

```r
install.packages("tidyverse")
library(tidyverse)
```

### Import Data

```r
bookings_df <- read_csv("hotel_bookings.csv")
```

### Inspect & Clean Data

```r
head(bookings_df)
colnames(bookings_df)
str(bookings_df)
```

Create another data frame using `bookings_df` that focuses on the average daily rate, which is referred to as `adr` in the data frame, and  `adults`

```r
new_df <- select(bookings_df, `adr`, adults)
```

To create new variables in your data frame, you can use the `mutate()` function. This will make changes to the data frame, but not to the original data set you imported. That source data will remain unchanged. 

mutate(new_df, total = `adr` / adults)
